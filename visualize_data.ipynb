```python
import matplotlib.pyplot as plt
import numpy as onp
import pandas as pd 

from utils import standardize_data
```

```python
DATA_PATH = "/Users/johnnyxi/Documents/phd/CausalDiscover/causal-velocity/data/synthetic/velocity"

meta = pd.read_csv(DATA_PATH + "/pairmeta.txt",  delim_whitespace=True,
                            header=None,
                            names=['pair_id', 'cause_start', 'cause_end', 'effect_start', 'effect_end', 'weight'],
                            index_col=0)

plt.figure(figsize=(9,6))

idx = [1,2,3,4,5,6]


for i, id in enumerate(idx):
    id = id - 1 
    pair_id = meta.index[id].astype(int)
    cause = meta["cause_start"].values[id].astype(int)
    effect = meta["effect_start"].values[id].astype(int)
    weight = meta["weight"].values[id]
    dat = onp.loadtxt(DATA_PATH + f"/pair{pair_id:04d}.txt")
    dat = dat[:1000]
    dat = standardize_data(dat, return_statistics = False, trim_outliers = 0)  ## optionally, trim off extreme points as outliers
    x_data = dat[:,0]
    y_data = dat[:,1]
    title = "x->y" if cause == 1 else "y->x"
    plt.subplot(3,2,i+1)
    plt.scatter(x_data, y_data, s=1)
    plt.title(title, fontsize = 20)
    plt.xticks([], [])
    plt.yticks([], [])

plt.figure(figsize=(9,6))


for i, id in enumerate(idx):
    id = id - 1 
    pair_id = meta.index[id].astype(int)
    cause = meta["cause_start"].values[id].astype(int)
    effect = meta["effect_start"].values[id].astype(int)
    weight = meta["weight"].values[id]
    dat = onp.loadtxt(DATA_PATH + f"/pair{pair_id:04d}.txt")
    dat = dat[:1000]
    dat = standardize_data(dat, return_statistics = False, trim_outliers = 0)  ## optionally, trim off extreme points as outliers
    x_data = dat[:,0]
    y_data = dat[:,1]
    plt.subplot(3,2,i+1)
    plt.hist(x_data, bins=50, density=True, alpha=0.5, label='x', color='blue') if cause == 1 else plt.hist(y_data, bins=50, density=True, alpha=0.5, label='y', color='blue')
    plt.title("cause", fontsize = 20)
    plt.xticks([], [])
    plt.yticks([], [])
```
